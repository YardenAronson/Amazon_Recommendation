# Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data cleaning

```python
df = pd.read_csv('/content/Amazon_Recommendation/amazon.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df['discounted_price'] = df['discounted_price'].str.replace('₹', '').str.replace(',','').astype(float)
df['actual_price'] = df['actual_price'].str.replace('₹', '').str.replace(',','').astype(float)
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '').astype(float) / 100
df['rating'] = df['rating'].str.replace('|','').replace('', np.nan).astype(float)
df['rating_count'] = df['rating_count'].str.replace(',','').astype(float)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
users = set()
for _, row in df[['user_id', 'user_name']].iterrows():
  ids = row['user_id'].split(',')
  names = row['user_name'].split(',')
  for i in range(len(ids)):
    users.add((ids[i],names[i]))
users = pd.DataFrame(users, columns=['user_id', 'user_name'])
users = users.drop_duplicates()
```

```python
users
```

```python
print(users.columns)
```

```python
# Drop duplicate user_ids before setting the index
users = users.drop_duplicates(subset='user_id')

# Set 'user_id' as the index
users.set_index('user_id', inplace=True, verify_integrity=True)

# Convert DataFrame to dictionary with user_id as keys
users_dict = users.to_dict(orient='index')
users_dict
```

```python
products = df[['product_id', 'product_name', 'img_link', 'product_link', 'category']].drop_duplicates()
products['category'] = products['category'].str.split('|')
# Ensure 'product_id' is unique before setting it as index
products.drop_duplicates(subset='product_id', inplace=True)  # Drop rows with duplicate product_id

# Set 'product_id' as the index
products.set_index('product_id', inplace=True)

# Convert DataFrame to dictionary with product_id as keys
products = products.to_dict(orient='index')
products
```

```python
rating = pd.DataFrame(columns=['user_id', 'product_id', 'rating','rating_count'])
for _, row in df[['user_id','rating','rating_count','product_id']].iterrows(): # Use a list to select multiple columns
  ids = row['user_id'].split(',')
  for id in ids:
    rating = pd.concat([rating, pd.DataFrame({'user_id': [id],
                                           'product_id': [row['product_id']],
                                           'rating': [row['rating']],
                                           'rating_count': [row['rating_count']]})],
                       ignore_index=True)
```

```python
rating
```

# EDA

```python
# Rating distribution
plt.figure(figsize=(8, 6))
sns.histplot(rating['rating'], bins=10, kde=True)
plt.title('Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

# Saving Database as CSV

```python
import json
with open('products.json', 'w') as f:
    json.dump(products, f)
with open('users.json', 'w') as f:
    json.dump(users_dict, f)
rating.to_csv('rating.csv', index=False)
```
